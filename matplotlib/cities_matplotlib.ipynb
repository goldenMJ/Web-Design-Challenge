```python
#dependencies 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load file to df
df = pd.read_csv('/Users/seve/Desktop/Web-Design-Challenge/Resources/cities.csv')
df.to_html()
```

```python
df.head()
```

```python
#city latitude(x) vs. Max Temperature(y)
x = df['Lat']
y = df['Max Temp']
plt.xlabel('latitude');plt.ylabel('Max Temperature')
plt.title("City Latitude vs. Max Temperature", fontsize=13.5)
plt.scatter(x,y)
```
